```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files


uploaded = files.upload()
if uploaded:
    img_path = next(iter(uploaded))
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
else:
    raise Exception("No image uploaded!")


img = cv2.resize(img, (256, 256))


mean_filtered     = cv2.blur(img, (5, 5))
median_filtered   = cv2.medianBlur(img, 5)
gaussian_filtered = cv2.GaussianBlur(img, (5, 5), sigmaX=1.0)
bilateral_filtered = cv2.bilateralFilter(img, d=9, sigmaColor=75, sigmaSpace=75)


def low_pass_filter(image, cutoff=30):
    f = np.fft.fft2(image)
    fshift = np.fft.fftshift(f)
    rows, cols = image.shape
    crow, ccol = rows // 2 , cols // 2
    mask = np.zeros_like(image, dtype=np.float32)
    mask[crow-cutoff:crow+cutoff, ccol-cutoff:ccol+cutoff] = 1
    fshift_filtered = fshift * mask
    img_back = np.fft.ifft2(np.fft.ifftshift(fshift_filtered))
    return np.abs(img_back)

def high_pass_filter(image, cutoff=30):
    f = np.fft.fft2(image)
    fshift = np.fft.fftshift(f)
    rows, cols = image.shape
    crow, ccol = rows // 2 , cols // 2
    mask = np.ones_like(image, dtype=np.float32)
    mask[crow-cutoff:crow+cutoff, ccol-cutoff:ccol+cutoff] = 0
    fshift_filtered = fshift * mask
    img_back = np.fft.ifft2(np.fft.ifftshift(fshift_filtered))
    return np.abs(img_back)

low_pass_filtered = low_pass_filter(img, cutoff=30)
high_pass_filtered = high_pass_filter(img, cutoff=30)


titles = ["Original", "Mean", "Median", "Gaussian", "Bilateral", "Low-Pass", "High-Pass"]
images = [img, mean_filtered, median_filtered, gaussian_filtered, bilateral_filtered, low_pass_filtered, high_pass_filtered]

plt.figure(figsize=(12, 8))
for i in range(len(images)):
    plt.subplot(2, 4, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
